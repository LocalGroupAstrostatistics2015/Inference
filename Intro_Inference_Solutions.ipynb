# Orbits in an elliptical disk: Theory

## Equipotential contours of standard model

```python
def elliptical_potential(R,phi,epsilon0=0.05,phib=0.,alpha=0.,p=0.,vc=1.):
    """Elliptical Disk potential: phi and phib in radian"""
    if alpha == 0.:
        axipot= vc**2.*numpy.log(R)
    else:
        axipot= vc**2./2./alpha*R**(2.*alpha)
    return axipot+epsilon0*vc**2./2.*R**p*numpy.cos(2.*(phi-phib))
```

```python
xx, yy= numpy.meshgrid(numpy.linspace(1.1,16.,101),numpy.linspace(0.,2.*numpy.pi,101),indexing='ij')
ax= axes([0.1,0.1,0.8,0.8],polar=True)
contour(yy,xx,elliptical_potential(xx,yy,epsilon0=0.15,phib=numpy.pi/2.),color='k',levels=numpy.log(numpy.linspace(2,14,7)))
```

## Check $\tilde{R}$

```python
def Rtilde(R,phi,epsilon0=0.05,phib=0.):
    return R*numpy.sqrt(((1-epsilon0)**2.*numpy.sin(phi-phib)**2.+numpy.cos(phi-phib)**2.)\
                        /((1-epsilon0)**2.*numpy.sin(phib)**2.+numpy.cos(phib)**2.))
```

```python
xx, yy= numpy.meshgrid(numpy.linspace(1.1,16.,101),numpy.linspace(0.,2.*numpy.pi,101),indexing='ij')
ax= axes([0.1,0.1,0.8,0.8],polar=True)
contour(yy,xx,Rtilde(xx,yy,epsilon0=0.15,phib=numpy.pi/2.),levels=numpy.linspace(2,14,7))
```

# The axisymmetric metallicity trend

## Likelihood function for linear fit with fixed $\Delta$

```python
def loglike_axi_linear(params,R,metals,Delta=0.2):
    """Log likelihood for the axisymmetric, linear metallicity trend: m(R/kpc-8)+b"""
    m= params[0]
    b= params[1]
    return -0.5*numpy.sum((metals-m*(R-8.)-b)**2.)/Delta**2.
```

## Contours of the likelihood function for linear fit with fixed $\Delta$

```python
ms= numpy.linspace(-0.3,0.3,51)
bs= numpy.linspace(-0.2,0.2,51)
index= (numpy.fabs(rcdata['RC_GALZ']) < 0.25)*(rcdata['RC_GALR'] > 4.)*(rcdata['RC_GALR'] < 12.)
logl= numpy.array([[loglike_axi_linear([m,b],rcdata['RC_GALR'][index],rcdata['METALS'][index]) for b in bs] for m in ms])
CS= contour(-2.*(logl.T-numpy.nanmax(logl)),origin='lower',extent=[ms[0],ms[-1],bs[0],bs[-1]])
clabel(CS, inline=1, fontsize=10)
xlabel(r'$m\,(\mathrm{dex\,kpc}^{-1})$')
ylabel(r'$b\,(\mathrm{dex}$')
```

# Simple elliptical-disk likelihood

## Likelihood function for the model with $\alpha = 0$ and $p=0$ with fixed axisymmetric trend

```python
def loglike_nonaxi_flatflat(params,R,phi,metals,Delta=0.2,alpha=0.,p=0.,m=-0.08,b=0.07):
    """Log likelihood for the flat-rotation curve, flat triaxiality elliptical-disk orbits model"""
    epsilon0= params[0]
    phib= params[1]
    Rt= Rtilde(R,phi,epsilon0=epsilon0,phib=phib)
    return -0.5*numpy.nansum((metals-m*(Rt-8.)-b)**2.)/Delta**2.
```

## Contours of the likelihood function for the model with $\alpha = 0$ and $p=0$ with fixed axisymmetric trend

```python
epsilon0s= numpy.linspace(0.,0.3,21)
phibs= numpy.linspace(-numpy.pi/2.,numpy.pi/2.,21)
index= (numpy.fabs(rcdata['RC_GALZ']) < 0.25)*(rcdata['RC_GALR'] > 4.)*(rcdata['RC_GALR'] < 12.)
logl= numpy.array([[loglike_nonaxi_flatflat([e,p],rcdata['RC_GALR'][index],rcdata['RC_GALPHI'][index],
                                            rcdata['METALS'][index]) for p in phibs] for e in epsilon0s])
CS= contour(-2.*(logl.T-numpy.nanmax(logl)),origin='lower',extent=[epsilon0s[0],epsilon0s[-1],phibs[0],phibs[-1]])
clabel(CS, inline=1, fontsize=10)
xlabel(r'$\epsilon_0$')
ylabel(r'$\phi_b$')
```

```python
print(epsilon0s[numpy.unravel_index(numpy.nanargmax(logl),(len(epsilon0s),len(phibs)))[0]], \
    phibs[numpy.unravel_index(numpy.nanargmax(logl),(len(epsilon0s),len(phibs)))[1]])
```

## Contours of the likelihood function for the model with $\alpha = 0$ and $p=0$ with fixed axisymmetric trend in terms of better parameters

A better set of parameters is $c_0 \equiv \epsilon_0\,\cos 2 \phi_b$ and $s_0 \equiv \epsilon_0\,\sin 2 \phi_b$. The likelihood is then

```python
def loglike_nonaxi_flatflat_improvedparams(params,R,phi,metals,Delta=0.2,alpha=0.,p=0.,m=-0.08,b=0.07):
    """Log likelihood for the flat-rotation curve, flat triaxiality elliptical-disk orbits model"""
    epsilon0= numpy.sqrt(params[0]**2.+params[1]**2.)
    phib= numpy.arctan2(params[1],params[0])/2.
    Rt= Rtilde(R,phi,epsilon0=epsilon0,phib=phib)
    return -0.5*numpy.nansum((metals-m*(Rt-8.)-b)**2.)/Delta**2.
```

```python
c0s= numpy.linspace(-0.3,0.3,21)
s0s= numpy.linspace(-0.3,0.3,21)
index= (numpy.fabs(rcdata['RC_GALZ']) < 0.25)*(rcdata['RC_GALR'] > 4.)*(rcdata['RC_GALR'] < 12.)
logl= numpy.array([[loglike_nonaxi_flatflat_improvedparams([c,s],rcdata['RC_GALR'][index],rcdata['RC_GALPHI'][index],
                                            rcdata['METALS'][index]) for s in s0s] for c in c0s])
CS= contour(-2.*(logl.T-numpy.nanmax(logl)),origin='lower',extent=[c0s[0],c0s[-1],s0s[0],s0s[-1]])
clabel(CS, inline=1, fontsize=10)
xlabel(r'$c_0 \equiv \epsilon_0\,\cos 2 \phi_b$')
ylabel(r'$s_0 \equiv \epsilon_0\,\sin 2 \phi_b$')
```

```python
print(c0s[numpy.unravel_index(numpy.nanargmax(logl),(len(c0s),len(s0s)))[0]], \
    s0s[numpy.unravel_index(numpy.nanargmax(logl),(len(c0s),len(s0s)))[1]])
```

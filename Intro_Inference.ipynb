# Orbits in an elliptical disk: Theory

We follow [Kuijken & Tremaine (1994)](http://adsabs.harvard.edu/abs/1994ApJ...421..178K) and consider an elliptical perturbation to a simple axisymmetric potential of the form:

$\Phi(R,\phi) = \Phi_0(R) + \Psi(R)\,\cos 2\,(\phi-\phi_b)$

where the axisymmetric part of the potential is given by a power law:

$\Phi_0(R) = \frac{V_c^2}{2\,\alpha}\,\left(\frac{R}{R_0}\right)^{2\alpha},\qquad \alpha \neq 0$

or

$\Phi_0(R) = V_c^2\,\ln R,\qquad \alpha = 0$ (flat rotation curve)

A simple form for the elliptical perturbation also has a power-law radial profile:

$\Psi(R) = \Psi_0\,\left(\frac{R}{R_0}\right)^p$

For this potential, the closed orbits are approximately elliptical with axis ratio $q_\Psi \equiv 1-\epsilon_\Psi + O(\epsilon_\Psi^2)$, where the dimensionless ellipticity $\epsilon_\Psi(R)$ is given by

$\epsilon_\Psi(R) = \frac{2\,\Psi(R)}{V_c^2(R)}\,,$

where $V_c(R)$ is the circular velocity. We also define $\epsilon_0 \equiv \epsilon_\Psi(R_0)$ and assume $\epsilon_0 \geq 0$, $-\pi/2 \leq \phi_b \leq \pi/2$. The equation for an ellipse in polar coordinates is

$r(\phi) = R\,\frac{(1-\epsilon_0)}{\sqrt{(1-\epsilon_0)^2\,\sin^2[\phi-\phi_b]+\cos^2[\phi-\phi_b]}}\,.$

Here is a useful website: [Wikipedia](http://en.wikipedia.org/wiki/Ellipse).

The simplest model is one where both $\alpha$ and $p$ are zero, that is, an axisymmetric background potential with a flat rotation curve combined with an elliptical perturbation for which the ellipticity does not change as a function of radius.

As an exercise, plot the equipotential contours of this potential.

Our model for the metallicity of the RC stars above is that the metallicity is constant along ellipses in $(R,\phi)$ with ellipticity $\epsilon_\Psi(R)$, with a linear trend in $[\mathrm{Fe/H}](R,0)$ and a constant metallicity dispersion. Therefore, the metallicity of a star at $(R,\phi)$ is given by

$[\mathrm{Fe/H}](R,\phi) = \left(m\,[\tilde{R}/\mathrm{kpc}-8]+b\right)+\delta+\mathrm{noise}\qquad (1)$

where $\tilde{R}$ is the radius along the ellipse through $(R,\phi)$ that goes through $\phi = 0$ and is therefore given by

$\tilde{R} = R\,\sqrt{\frac{(1-\epsilon_0)^2\,\sin^2[\phi-\phi_b]+\cos^2[\phi-\phi_b]}{(1-\epsilon_0)^2\,\sin^2[\phi_b]+\cos^2[\phi_b]}}\,.$

Check that this equation is correct by writing a function that returns $\tilde{R}$ and contouring it.

This simple model has the following free parameters: (a) the parameters of the axisymmetric metallicity trend: $(m,b)$ and the metallicity dispersion $\Delta$, (b) the parameters of the ellipticity: the basic parameters $\epsilon_0$ and $\phi_b$, and the radial profile specified by $p$. The rotation curve has the free parameters $V_c$ and $\alpha$, but we shall keep those fixed throughout this exercise (why is this?).
